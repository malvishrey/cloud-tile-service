```python
import subprocess
import datetime
planet_scope = {}
# Define the date range
start_date = datetime.date(2022, 12, 1)
end_date = datetime.date(2023, 4, 30)
# start_date = 
# Define the GeoJSON file and other parameters
geojson_file = "huc10_bvc_simplified.geojson"
command_base = "pcf"

# Iterate through each date in the range
current_date = start_date
while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")
    
    # Build the command
    command = [command_base, geojson_file, date_str, "-x","-t","90"]
    
    # Execute the command
    result = subprocess.run(command, capture_output=True, text=True)
    print(result)
    
    # Check the output for the URL
    if "SUCCESS" in result.stdout:
        print(f"Date: {date_str}, URL found:")
        for line in result.stdout.splitlines():
            if "XYZ:" in line:
                print(line.strip())
                planet_scope[date_str] = line.strip()
    
    # Move to the next day
    current_date += datetime.timedelta(days=1)
    # break
```

```python
for date, value in planet_scope.items():
    if "XYZ:" in value:
        # Extract the URL after 'XYZ:'
        planet_scope[date] = value.split("XYZ:")[1].strip()

# Print the updated dictionary
(planet_scope)
```

```python
import json
with open('ps_daily.json', 'w') as json_file:
    json.dump(planet_scope, json_file, indent=4)
```
